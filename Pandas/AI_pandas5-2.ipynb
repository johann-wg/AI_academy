# TIF 실습

```python
tif = pd.read_csv('TIF.csv')
tif
```

```python
tif['MEASUREDATE'] = pd.to_datetime(tif['MEASUREDATE'])
```

```python
tif_pivot = tif.pivot_table('VALUE', index=tif['MEASUREDATE'], columns=tif['MEASURETYPE'])
tif_pivot
```

```python
tif_pivot_result = tif_pivot.loc[:, ['humi','vibr','powe','temp','tira','nois']]
tif_pivot_result
```

```python
# 1분단위 MA
tif_ma = tif_pivot_result.rolling('1min').mean()
tif_ma.plot()
```

```python
# corr() 센서들의 상관도 top3
tif_corr = tif_ma.corr()
sns.heatmap(tif_corr.corr())
plt.show()
```

```python
# top3 뽑기
tif_corr.unstack().abs().sort_values()
```

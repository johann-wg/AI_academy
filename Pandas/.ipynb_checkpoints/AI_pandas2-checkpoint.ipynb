```python
sr = pd.Series([0,1,np.nan,3,4])
sr
```

```python
# 결측치 확인 함수
print(sr.isna())
print(sr.isnull())
```

```python
sr.notna()
sr.notnull()
```

```python
# 결측치 합계
sr.isna().sum()
```

```python
# 결측치 존재하는 행 찾기
df = pd.DataFrame({'A':[1,2,None,4,None],
                   'B':[None, 6, 7, None, 9]})
df_with_nan = df.isna()
mask = df_with_nan.any(axis=1)
result = df[mask]
print(result)
```

```python
# 결측치 채우기
sr = pd.Series([0, 1, np.nan, np.nan, 4])
sr.fillna(method='ffill') # forword fill 앞의 것을 채운다
sr
```

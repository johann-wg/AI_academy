# 다운샘플링 : 
# 기존 데이터의 샘플링 개수를 감소시키는 것

```python
data = {'날짜':['2023-05-01', '2023-05-02', '2023-05-03'],
        '판매량':[10,15,12]}
df = pd.DataFrame(data)
df
```

# 이건 외워둬야 함

```python
df['날짜'] = pd.to_datetime(df['날짜'])
df = df.set_index('날짜')
df
```

```python
downsampled_df = df.resample('W').sum()
downsampled_df
```

```python
downsampled_df = df.resample('W-MON').sum() # 월요일을 시작일로
downsampled_df
```

# 업샘플링:
## 기존 데이터의 샘플링 개수를 증가시키는 것

```python
upsampled_df = df.resample('H').sum()
upsampled_df
```

# interpolate 보간

```python
data = {'date':pd.date_range(start='2023-05-01', end='2023-05-14', freq='D'),
        'value':[1,6,4,None,36,17,None,91,180,23,88,12,44,23]}
df = pd.DataFrame(data)
df
```

```python
df.set_index('date', inplace=True)
df
```

```python
# downsampled_df = df.resample('2D')
downsampled_df = df.interpolate(method='linear')
downsampled_df
```

# 롤링 윈도우 Rolling Window

```python
data = {'date':pd.date_range(start='2023-05-01', end='2023-05-10', freq='D'),
        'value':[1,6,4,12,36,17,44,91,180,23]}
df = pd.DataFrame(data)
df.set_index('date', inplace=True)
```

```python
rolling_mean = df['value'].rolling(window=5).sum()
rolling_mean
```

```python
rolling_mean = rolling_mean = df['value'].rolling(window=5)
for i in rolling_mean:
    print(i)
```

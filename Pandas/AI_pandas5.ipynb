```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('COVID19.csv', encoding='cp949')
df
```

```python
df.columns = ['Category', 'Type', 'Date', 'Headcount', 'Unnamed']
df
```

```python
type(df['Date'][0])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(type(df['Date'][0]))
df
```

```python
df = df.set_index('Date')
df
```

```python
df.drop(['Unnamed'], axis=1, inplace=True)
df
```

```python
df.isna().sum()
```

```python
print(len(df[df.isna().any(axis=1)]))
df[df.isna().any(axis=1)]
```

```python
# 결측치가 있는 행만 지우는 것
df = df.dropna(axis=0)
df
```

```python
df['Headcount'].sum()
```

```python
df.groupby('Type')['Headcount'].sum()
df.groupby('Category')['Headcount'].sum()
```

```python
df.pivot_table(index='Category', columns='Type', values='Headcount').sum()
df
```

```python
df_2020 = df.loc['2020-04-09':'2020-12-31']
df_2020
df_2021 = df.loc['2021-01-01':'2021-12-31']
df_2021
```

```python
df.describe()
```

```python
pivot_table = df.pivot_table(index='Category', columns='Type', values='Headcount', aggfunc='max')
pivot_table
```

# 교차일람표

```python
cross_tap = pd.crosstab(index=df['Category'], columns=df['Headcount'])
cross_tap
```

```python
# 음의 값 제거하기
df = df.loc[df['Headcount'] >= 0]
df
```

```python
cross_tap = pd.crosstab(index=df['Category'], columns=df['Headcount'])
cross_tap
```

# 다운샘플링

```python
downsampled_df = df.resample('M').sum(numeric_only=True)
downsampled_df
```

```python
downsampled_df.plot()
```

```python
rolling_mean = downsampled_df['Headcount'].rolling(window=3).sum()
rolling_mean.plot()
```

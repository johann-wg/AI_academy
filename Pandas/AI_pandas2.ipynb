```python
# 결측치 존재하는 행 찾기
df = pd.DataFrame({'A':[1,2,None,4,None],
                   'B':[None, 6, 7, None, 9]})
df_with_nan = df.isna()
mask = df_with_nan.any(axis=1)
result = df[mask]
print(result)
```

```python
# 결측치 채우기
sr = pd.Series([0, 1, None, None, 4])
sr = sr.fillna(method='ffill') # forword fill 앞의 것을 채운다 / back fill
sr
```

# 오후

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({'food':['bacon','pulled pork','bacon','Pastrami','corned beef',
                              'Bacon', 'pastrami','honey ham', '베이컨'],
                     'ounces':[4,3,12,6,7.5,8,3,5,7]})
data
```

```python
# 모두 소문자로
data['food'].str[:3]
data['food'] = data['food'].str.lower()
# .str 하면 파이썬에서 사용하는 str클래스 메서드 모두 사용 가능
```

```python
data
```

### inplace=True는 원본 자체가 바뀜 / 반환이 없음 그래서 변수할당 필요없음
### 파이썬은 변수 할당이 느림. 

```python
# data = data.replace('베이컨', 'bacon')
# inplace=True는 원본 자체가 바뀜 / 반환이 없음 그래서 변수할당 필요없음
# 파이썬은 변수 할당이 느림. 
data['food'].replace('베이컨', 'bacon', inplace=True) 
data
```

# 파이썬 map과 판다스 map은 다른 것
# map은 시리즈에서

```python
# 열 추가하기
meat_dict = {'bacon':'pig', 'pulled pork':'pig', 'pastrami':'cow', 'corned beef':'cow',
             'honey ham':'pig'}

data['animal'] = data['food'].map(meat_dict)
data
```

```python
data['food'].map(lambda x: meat_dict[x.lower()])
# 새로운 시리즈로
```

```python
# 메서드 체이닝
data['food'].str.lower().map(meat_dict)
```

# replace

# apply

```python
df = pd.DataFrame(np.random.randn(4,3), columns=['a', 'b', 'c'], index=list('ABCD'))
df
```

```python
df.apply(lambda x: x.mean())
```

```python
df.apply(lambda x: x.mean(), axis=1)
df.apply(lambda x: x.mean(), axis='columns')
```

```python
# 데이터 프레임에도 쓸 수 있는 map
df.applymap(lambda x: f'{x:.2f}')
```

```python
df.sort_index(axis=1, ascending=False)
```

```python
df.sort_values(by=['a', 'b'], ascending=False)
# a기준으로 먼저 정렬 후 같은 게 있으면 b를 기준으로 정렬
```

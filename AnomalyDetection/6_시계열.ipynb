```python
import pandas as pd

df = pd.read_csv('data.csv')
df
```

```python
df.index = pd.to_datetime(df['일시'])
df.drop('일시', axis=1, inplace=True)
df
```

```python
df_resample = df.resample('1s').mean()
df_resample
```

```python
df_resample_interpolated = df.resample('1s').mean().interpolate()
```

```python
df_resample_interpolated.iloc[:50000].plot(backend='plotly')
```

```python
df_resample_interpolated.iloc[:50000].interpolate(method='polynomial', order=2).plot(backend='plotly')
```

```python
df_resample_interpolated.iloc[:50000].interpolate().rolling('120s').mean().plot(backend='plotly')
```

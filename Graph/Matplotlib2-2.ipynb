# 박스플롯
* 통계적인 특징. 정규분포와 비슷, 이상치는 0.3%

```python
iris = load_iris()
x = iris['data']

df = pd.DataFrame(x, columns=iris.feature_names)

# boxprops = dict(facecolor='lightblue', edgecolor='black') boxplot의 인자로 넣으면 된다.

plt.figure()
plt.boxplot(df, labels=df.columns, notch=True, patch_artist=True, 
            boxprops={'facecolor':'green', 'edgecolor':'magenta'})

plt.show()
```

# 기준선 긋기(시계열에서 자주 쓰임)

```python
r1 = np.random.rand(10)
r2 = np.random.rand(10)
plt.figure()
plt.plot(r1)
plt.plot(r2, color='red')

plt.axhline(0.5, color='red', alpha=0.5, xmin=0.2, xmax=0.8)
plt.axvline(4, color='green', alpha=0.5, ymax=0.6)

plt.axvspan(5,8, color='magenta', alpha=0.5, hatch='/', fill=False)
plt.axhspan(0.2,0.4, xmin=0.2, xmax=0.5, color='orange', alpha=0.3)

plt.fill_between(np.arange(2, 7), r1[2:7], r2[2:7],alpha=0.5, color='green')
plt.show()
```

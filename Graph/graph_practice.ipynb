```python
plt.style.use('classic')

fig = plt.figure()
ax = fig.add_subplot(111)

data = [0, 1]
ax.plot(data)

ax.set_title('My graph', 
             fontdict={'fontsize':30, 'fontweight':'bold'},
             loc='right',
             color='green') # 제목 지정 함수

ax.set_xlabel('X Axis', fontdict={'fontsize':22, 'fontweight':'bold'}, loc='left', color='blue') 
ax.set_ylabel('Y Axis', fontdict={'fontsize':22, 'fontweight':'bold'}, loc='bottom', color='red')

ax.set_xticks([1,2,3], [1,2,3])
ax.set_yticks([10,20,30])

# ax.set_xticklabels(['A','B','C'])
# ax.set_yticklabels(['D','E','F'])

plt.show()
```

```python
plt.style.use('ggplot')

base_url = ('https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/')

anime_sotck_return_csv = os.path.join(base_url, 'anime_stock_returns.csv')

df = pd.read_csv(anime_sotck_return_csv, index_col=0)

fig = plt.figure(figsize=(16,8))

ax = fig.add_subplot(111)

ax.set_title('애니메이션 주가등락률', fontsize=30, color='green', fontweight='bold')

x_tick = range(0, len(df), 30)
x_label = [df.index[i] for i in x_tick]

ax.set_xticks(x_tick, x_label, rotation=60)

ax.plot(df.index, df['TOEI ANIMATION'], color='red', label='TOEI')
ax.plot(df.index, df['IG Port'], color='blue', label='IG')

plt.legend(fontsize=15, loc='upper left')
plt.show()
```

```python
plt.style.use('ggplot')

iris = load_iris()

data = iris.data
target = iris.target
feature_names = iris.feature_names

# colors = np.arange(150)

fig = plt.figure(figsize=(16,8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.scatter(data[:,0], data[:,1], c=target)
ax1.set_xlabel(feature_names[0])
ax1.set_ylabel(feature_names[1])
ax1.set_title('Sepal Lenght VS Sepal Width')

ax2.scatter(data[:,2], data[:,3], c=target)
ax2.set_xlabel(feature_names[2])
ax2.set_ylabel(feature_names[3])
ax2.set_title('Petal Lenght VS Petal Width')


plt.show()
```

```python

tips = sns.load_dataset('tips')

#Figure 생성
fig = px.Figure()

#그래프 생성
fig.add_trace(go.Scatter(x=tips['total_bill'], y=tips['tip'], mode='markers'))


# 버튼 2개 생성
fig.update_layout(
    updatemenus=[
        dict(
            type="dropdown",
            direction="down",
            buttons=list([
                dict(args=["type", "scatter"],  # 1번 버튼
                     label="Scatter Plot",
                    method="animate"),
                dict(args=["type", "bar"],  # 2번 버튼
                    label="Bar Chart",
                    method="restyle")
            ]),
        ),
    ]
)

fig.show()
```

```python
# 데이터 불러오기
flights = sns.load_dataset('flights')

fig = px.line(flights, x="month", y="passengers", facet_col='year')

fig.show()
```

```python
#데이터 불러오기
iris = sns.load_dataset('iris')

grouped = iris.groupby('species').mean().reset_index()

#그래프 그리기
fig = px.line_polar(grouped, r='sepal_length', theta='species', color='species')

# 내부 색칠하기
fig.update_traces(fill='toself')

fig.show()
```

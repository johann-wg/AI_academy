```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 서식 지정

```python
fig = plt.figure()

ax = fig.add_subplot(111)

x = [0,1]
y = [0,1]

ax.plot(x,y)

plt.show()
```

```python
fig = plt.figure(figsize=(6,6), dpi=100, # 해상도
                 facecolor='green', # 테두리 색상
                 edgecolor='m', # 테두리 색상
                 linewidth=5, # 테두리 두께
                 frameon=True) # 테두리 표시 여부

ax = fig.add_subplot(111, facecolor='black',
                     xlim=[0, 10],
                     ylim=[0, 10000])

x = [0,1]
y = [0,1]

ax.plot(x,y)


plt.show()
```

```python
fig = plt.figure(figsize=(6,6), dpi=100, # 해상도
                 facecolor='green', # 테두리 색상
                 edgecolor='m', # 테두리 색상
                 linewidth=5, # 테두리 두께
                 frameon=True) # 테두리 표시 여부

x = [0,9]
y = [0,9999]

colors = ['black', 'cyan', 'yellow', 'blue']

for i in range(1,5):
    ax = fig.add_subplot(2,2,i, facecolor=colors[i-1], xlim=[0, 10], ylim=[0, 10000])
    ax.plot(x,y)
    fig.subplots_adjust(wspace=0.5, hspace=0.5) # 레이아웃 조절

plt.show()
```

# 선의 굵기와 색상, 범례

```python
x = np.array([0, 1])
y = [0, 1]

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, xlim=(-0.2, 4))

ax.plot(x,y, linewidth=1, color='b', label='linewidth=1(alpha=1)')
ax.plot(x + 0.1,y, linewidth=2, color='r', alpha=0.5,label='linewidth=2(alpha=0.5)')
ax.plot(x + 0.2,y, linewidth=3, color='g', alpha=0.3,label='linewidth=3(alpha=0.3)')

plt.legend()
plt.show()
```

# 선의 끝 모양

```python
x = [0.1, 1]
y = np.array([0, 0])

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, xlim=(0, 1.7), ylim=(-0.2, 1.2))

ax.plot(x,y + 1, linewidth=12, solid_capstyle='butt', label='butt')
ax.plot(x,y + 0.5, linewidth=12, solid_capstyle='round', label='round')
ax.plot(x,y, linewidth=12, solid_capstyle='projecting', label='projecting')


plt.legend(loc=0)
plt.show()
```

```python
x = np.array([0, 1,2,3])
y = [0, 1.0, 0.2, 1.2]

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, xlim=(0, 4.5))

ax.plot(x + 1, y, linewidth=10, solid_joinstyle='miter', label='miter')
ax.plot(x + 0.5, y, linewidth=10, solid_joinstyle='round', label='round')
ax.plot(x, y, linewidth=10, solid_joinstyle='bevel', label='bevel')



plt.legend(loc=0)
plt.show()
```

```python
x = np.array([0, 1])
y = np.array([1.9, 1.9])

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, xlim=(-0.1, 1.7), ylim=(0, 2))

ax.plot(x, y, linewidth=10, linestyle='solid', label='solid')
ax.plot(x, y - 0.1, linewidth=10, linestyle='-', label='-')

ax.plot(x, y-0.4, linewidth=10, linestyle='dashed', label='dashed')
ax.plot(x, y - 0.6, linewidth=10, linestyle='--', label='--')

ax.plot(x, y-0.9, linewidth=10, linestyle='dotted', label='dotted')
ax.plot(x, y - 1, linewidth=10, linestyle=':', label=':')

plt.legend()
plt.show()
```

# 산점도에서 점의 색과 불투명도 지정

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(x, y, label='sin', color='cyan', edgecolors='magenta')
ax.scatter(x, z, label='cos', color='red', edgecolors='black', alpha=0.3)

plt.legend(framealpha=0.5, loc=0)
plt.show()
```

# 컬러맵 사용하기

```python
import matplotlib.cm as cm
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

x = np.arange(20)
y = np.arange(20)

ax.scatter(x, y + 5, color=cm.plasma(0.1))
ax.scatter(x, y, color=cm.get_cmap('spring')(0.9))

plt.show()
```

# 텍스트 그리기

```python
fig = plt.figure()
ax = fig.add_subplot(111)

fps = {'family':'fantasy', 'style':'italic', 'weight':'heavy', 'size':40, 'color':cm.spring(0.7)}

ax.text(2, 6, 'Text', size=40, alpha=0.5, rotation=45, 
        fontdict=fps)

ax.set_xticks([i * 2 for i in range(6)])
ax.set_yticks([i * 2 for i in range(6)])

plt.show()
```

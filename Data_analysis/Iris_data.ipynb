```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.describe()
```

```python
iris['species'].value_counts()
```

```python
setosa = iris.loc[iris['species']=='setosa']
verosicolor = iris.loc[iris['species']=='verosicolor']
virginica = iris.loc[iris['species']=='virginica']
```

```python
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(221)
sns.histplot(setosa['sepal_length'], bins=10, ax=ax1, color='blue', label='Setosa', alpha=0.5)
sns.histplot(verosicolor['sepal_length'], bins=10, ax=ax1, color='green', label='verosicolor', alpha=0.5)
# sns.histplot(virginica['sepal_length'], bins=10, ax=ax1, color='red', label='virginica', alpha=0.5)
ax1.set_xlabel("Sepal Length(cm)")
ax1.legend()

ax2 = fig.add_subplot(222)
sns.histplot(setosa['sepal_length'], bins=10, ax=ax2, color='blue', label='Setosa', alpha=0.5)
sns.histplot(verosicolor['sepal_length'], bins=10, ax=ax2, color='green', label='verosicolor', alpha=0.5)
sns.histplot(virginica['sepal_length'], bins=10, ax=ax2, color='red', label='virginica', alpha=0.5)
ax2.set_xlabel("Sepal_width(cm)")
ax2.legend()

ax3 = fig.add_subplot(223)
sns.histplot(setosa['sepal_length'], bins=10, ax=ax3, color='blue', label='Setosa', alpha=0.5)
sns.histplot(verosicolor['sepal_length'], bins=10, ax=ax3, color='green', label='verosicolor', alpha=0.5)
sns.histplot(virginica['sepal_length'], bins=10, ax=ax3, color='red', label='virginica', alpha=0.5)
ax3.set_xlabel("Petal_length(cm)")
ax3.legend()


ax4 = fig.add_subplot(224)
sns.histplot(setosa['sepal_length'], bins=10, ax=ax4, color='blue', label='Setosa', alpha=0.5)
sns.histplot(verosicolor['sepal_length'], bins=10, ax=ax4, color='green', label='verosicolor', alpha=0.5)
sns.histplot(virginica['sepal_length'], bins=10, ax=ax4, color='red', label='virginica', alpha=0.5)
ax4.set_xlabel("Petal_width(cm)")
ax4.legend()

plt.show()



```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,10))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
colors = ['blue', 'red', 'green']

for i, feature in enumerate(features):
    row = i // 2
    col = i % 2
    ax = axs[row, col]

    for j, species in enumerate(iris['species'].unique()):
        sns.histplot(iris[iris['species'] == species][feature], 
                     bins=10, ax=ax, color=colors[j], kde=True, label=species)
    
    ax.set_xlabel(feature.capitalize() + '(cm)')
    ax.legend()
plt.show()
```

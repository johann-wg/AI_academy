```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_table('movielens/movies.dat', sep='::',
                    header=None, names=['MovieID', 'Title', 'Genres'],engine='python')

users = pd.read_table('movielens/users.dat', sep='::',
                    header=None, names=['UserID','Gender','Age','Occupation'],engine='python', index_col=False)

ratings = pd.read_table('movielens/ratings.dat', sep='::',
                    header=None, names=['UserID', 'MovieID', 'Rating', 'Timestamp'],engine='python')

```

```python
ratings = ratings.drop('Timestamp', axis=1)
```

# 병합 후 진행

```python
data = pd.merge(pd.merge(ratings, users), movies)
data
```

counts = data['Title'].value_counts()

```python
filter250 = data.groupby('Title').size()
# filter250[filter250 >= 250]
title250 = filter250.index[filter250 >= 250]
```

```python
mean_rating = data.pivot_table('Rating', index='Title', columns='Gender').loc[title250]
# data.loc[title250].pivot_table('Rating', index='Title', columns='Gender')
mean_rating
```

```python
mean_rating.sort_values(by='F', ascending=False)
```

```python
import pandas as pd
```

```python
dfs = []
for y in range(1880, 2011):
    df = pd.read_csv(f'babynames/yob{y}.txt', names=['name', 'gender', 'births'])
    df['year'] = y
    dfs.append(df)
names = pd.concat(dfs)
names
```

```python
# names['name'].map(lambda x: x[-1])
last_letter = names['name'].str[-1]
```

```python
last_names = names.pivot_table('births', index=last_letter, columns=['gender', 'year'], aggfunc='sum')
last_names
```

```python
F =last_names.loc[:, 'F']
top5_F = F.sum(axis=1).sort_values(ascending=False).index[:5]
```

```python
(F / F.sum()).loc[top5_F].T.plot()
```

```python
M =last_names.loc[:, 'M']
top5_M = M.sum(axis=1).sort_values(ascending=False).index[:5]
(M / M.sum()).loc[top5_M].T.plot()
```
